```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import cross_val_score
```

```python
x , y = make_regression(n_samples=100,n_features=1 , n_informative= 1 , n_targets= 1 ,noise=20)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(x,y,test_size=0.2 , random_state=2)
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
class GDRegrssor:

    def __init__(self,learning_rate,epochs):
        self.m = 100
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self,x,y):
        # Calculate b using gradien descent
        for i in range(self.epochs):
            loss_slop_b = -2 * np.sum ( y - self.m*x.ravel() - self.b)
            self.b = self.b - (self.lr * loss_slop_b)
            loss_slop_m = -2 * np.sum ( (y - self.m*x.ravel() - self.b) * x.ravel())
            self.m = self.m - (self.lr * loss_slop_m)
            # print(loss_slop,self.b)

        print(self.m , self.b)

    def predict(self,x):
        return self.m * x + self.b
```
