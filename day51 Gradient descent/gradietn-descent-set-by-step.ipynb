```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x, y = make_regression(n_samples=4 ,n_features=1 , n_informative= 1,noise=80 , random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# Lets apply gradient descent assuming is constant m=78.35
# and lets assume the starting for intercept b=0
y_pred = ((78.35 * x )+ 100).reshape(4)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label = 'OLS')
plt.plot(x,y_pred,color='#00a65a',label='b=0')
plt.legend()
plt.show()
```

```python
m= 78.35
b=100

loss_slop = -2 * np.sum ( y - m*x.ravel() - b)
loss_slop
```

```python
lr = 0.1

step_size = loss_slop * lr
step_size
```

```python
b = b-step_size
b
```

```python
y_pred1 =  ((78.35 * x) + b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label = 'OLS')
plt.plot(x,y_pred1,color='#00a65a',label='b={}'.format(b))
plt.plot(x,y_pred,color="#A3E4D7",label='b=0')
plt.legend()
plt.show()
```

```python
b = -100
m=78.35
lr = 0.01

epochs = 100

for i in range(epochs):
    loss_slop = -2 * np.sum( y - m*x.ravel() - b)
    b = b - (lr*loss_slop)

    y_pred = m*x +b
    plt.plot(x,y_pred)

plt.scatter(x,y)
```
