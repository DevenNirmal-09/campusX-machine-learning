```python
import requests

# Define the URL for the Spotify search API
url = "https://spotify23.p.rapidapi.com/search/?type=multi&offset=0&limit=10&numberOfTopResults=5"

# Set the headers with your RapidAPI key
headers = {
    "x-rapidapi-host": "spotify23.p.rapidapi.com",
    "x-rapidapi-key": "fb0d5fd166mshd80e248186356f2p1abad9jsn3ddad85c725f"  # Replace with your actual API key
}

# Send the GET request to the API with the headers
response = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Print the JSON response from the API
    print(response.json())
else:
    # Print an error message if the request failed
    print(f"Error: {response.status_code} - {response.text}")
```
